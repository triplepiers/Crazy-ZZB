# 成年人死亡率预测（统计建模）

## 0 实验要求

本项目旨在通过世界卫生组织所收集的所有国家健康状况及其他相关因素（人均酒精消费量、国家卫生支出、免疫覆盖率、BMI、GDP、人口、身体状况、收入信息和教育等），对各国的人口死亡率进行预测。该项研究将有助于人们了解哪些因素对死亡率有重要影响，从而指导人们能够采取有效措施，降低人口死亡率。

1. 训练数据包含2336条记录和22个字段，对训练数据进行一定的可视化数据分析
2. 利用训练数据，选择合适的信息作为特征建立回归模型
3. 利用MO平台进行模型性能评估

可以使用基于 Python 的 Pandas、Numpy、Sklearn 等库进行相关特征处理，使用 Sklearn 框架训练分类器，也可编写深度学习模型，使用过程中请注意 Python 包（库）的版本。
